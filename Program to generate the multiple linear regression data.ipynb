```python
import random
import csv

# Define number of data points
num_data_points = 100

# Define spending ranges
min_rd_spend = 10000
max_rd_spend = 150000
min_admin_spend = 5000
max_admin_spend = 30000
min_marketing_spend = 20000
max_marketing_spend = 100000

# Define states
states = ["CA", "NY", "WA", "TX"]

# Create empty lists to store data
data = []

# Generate random data points
for _ in range(num_data_points):
    rd_spend = random.randint(min_rd_spend, max_rd_spend)
    admin_spend = random.randint(min_admin_spend, max_admin_spend)
    marketing_spend = random.randint(min_marketing_spend, max_marketing_spend)
    state = random.choice(states)
    profit = 0.1 * rd_spend + 0.05 * admin_spend + 0.2 * marketing_spend + random.uniform(-10000, 10000)
    data.append([rd_spend, admin_spend, marketing_spend, state, profit])

# Define filename for the CSV file
filename = "transposed_multi_linear_regression_data.csv"

# Open the CSV file for writing in 'w' (write) mode with newline='' to avoid extra newlines
with open(filename, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    # Transpose the data before writing to achieve row-wise format
    writer.writerows(zip(*data))  # Use zip(*) for efficient transposition

print(f"Transposed data saved to CSV file: {filename}")
```

```python
import random
import csv

# Define number of data points
num_data_points = 100

# Define spending ranges
min_rd_spend = 10000
max_rd_spend = 150000
min_admin_spend = 5000
max_admin_spend = 30000
min_marketing_spend = 20000
max_marketing_spend = 100000

# Define states
states = ["CA", "NY", "WA", "TX"]

# Create empty lists to store data
rd_spend = []
admin_spend = []
marketing_spend = []
profit = []
state = []

# Generate random data points
for _ in range(num_data_points):
    rd_spend.append(random.randint(min_rd_spend, max_rd_spend))
    admin_spend.append(random.randint(min_admin_spend, max_admin_spend))
    marketing_spend.append(random.randint(min_marketing_spend, max_marketing_spend))
    state.append(random.choice(states))
    profit.append(0.1 * rd_spend[-1] + 0.05 * admin_spend[-1] + 0.2 * marketing_spend[-1] + random.uniform(-10000, 10000))

# Combine data into a single list for easier handling
data = list(zip(rd_spend, admin_spend, marketing_spend, state, profit))

# Define column headings
column_headings = ["R&D Spend", "Administration Spend", "Marketing Spend", "state", "Profit"]

# Define filename for the CSV file
filename = "multi_linear_regression_data.csv"

# Open the CSV file for writing in 'w' (write) mode with newline='' to avoid extra newlines
with open(filename, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    # Write the column headings as the first row
    writer.writerow(column_headings)
    # Write the data as subsequent rows
    writer.writerows(data)

print(f"Data with column headings saved to CSV file: {filename}")
```
